1. Resposta:

Listas: São estruturas mutáveis, delimitadas por colchetes[], ordenadas por índices inteiros que representam posição e capazes de conter diferentes tipos de informação consigo ao mesmo tempo(heterogêneas). Porém, a realização de operações na mesma é muito difícil, uma vez que as operações numéricas têm de ser realizadas dado a dado por um laço.

Dicionário: Assim como listas, são mutáveis e heterogêneas, no entanto não são ordenadas, sua indexação é feita por intermédio de Chaves específicas e são limitadas por chaves{}. Assim como as listas, também existe a dificuldade com operações numéricas.

Array numpy: Diferente das anteriores, não reside no python padrão, mas numa biblioteca a parte chamada numpy. Possui bastantes similaridades com listas, não possuem chaves como nos dicionários, mas tem o bônus de possibilitar que a realização de operações ocorra dentro do Array sem a necessidade de laços(operações vetorizadas),tornando o código mais eficiente. Além disso, tem a característica de homogeneizar os dados para um mesmo tipo dando prioridade nesta ordem: Str>complex>float>int. Bastando apenas um dado do tipo str para tornar todo o resto do array em str, por exemplo.

Séries Pandas: Possuem muitas semelhanças com o Array numpy, no entanto podem ser indexadas por índices específicos tais como as chaves em dicionários, apesar de manter a ordem em listas e numpy. Assim como no Array, e diferente do dicionário e das listas, existe uma obrigatoriedade das duas 'listas' em questão(No caso do series o Index e os dados. No numpy os dois arrays que se deseja trabalhar) para que essas tenham um mesmo tamanho, dessa forma podem ser realizadas tanto operações como a organização. No series também ocorre a questão da homogeneidade dos dados. E outro fator interessante do Series é sua saída onde se é possível visualizar claramente o Index(Que se apresenta como uma coluna ao lado dos dados, que também estão organizados de forma horizontal), que se difere das estruturas anteriores.
